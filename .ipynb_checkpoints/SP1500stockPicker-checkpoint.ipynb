```python
## First we import the required packages:
import numpy as np
from feature_selector import FeatureSelector
import pandas as pd
import datetime
import pickle
from sklearn.impute import SimpleImputer
import dateutil.relativedelta
```

```python
## We then load the required yield and financial ratios databse:
ratio= pd.read_csv('1990_2019.csv')
ratio = ratio.convert_objects(convert_numeric=True)
yields=pd.read_csv('yield_1962_2019.csv')
yields = yields.convert_objects(convert_numeric=True)
# We wanna use the datetime format to facilitate the next steps so let's change it right now:
ratio.public_date=pd.to_datetime(ratio.public_date)
yields.datadate=pd.to_datetime(yields.datadate)
```

```python
d = datetime.datetime.strptime("2000-01-01", "%Y-%m-%d")
d2= datetime.datetime.strptime("2019-01-31", "%Y-%m-%d")
ratio= ratio[ratio["public_date"].isin(pd.date_range(d, d2))]
yields= yields[yields["datadate"].isin(pd.date_range(d, d2))]
```

```python
ratio=ratio.reset_index(drop=True)
yields=yields.reset_index(drop=True)
```

#### Combining the datasets:
Now we want to combine both datasets on the columns "gvkey" and the date column.
For the financial ratios, the column "public_date" represents the date of publication of the financial report.
For the yields, the column ('datadate') represents the 

```python
import pandas as pd
from datetime import timedelta
from IPython.display import clear_output
import time
import numpy as np
import progressbar
### Price devient yields et BS devient ratio                                                                                                                                                                                                          
dict_new_entries = {}    # We are using a dictionary to append the list of rows                                                                                                                                                       
main=pd.DataFrame()      # Resulkting dataFrame converted from  dict_new_entries                                                                                                                                                      

yields_date = yields['datadate'] # Extract the columns we need. Much quicker for search/iterating                                                                                                                                       
yields_gvkey  = yields['gvkey']
ratio_gvkey     = ratio['gvkey']

number_columns = ratio.shape[1]           # Create empty row of the proper size. Will generate a row of NaNs                                                                                                                             
empty_row = [None] * number_columns
```

```python
for z in progressbar.progressbar(range(0,len(yields))):
# for z in progressbar.progressbar(range(0,1000)):                                                                                                                                                                                    
    # start = time.time()   ## For keeping track of iteration duration                                                                                                                                                                
    a = yields_date[z]
    b = yields_gvkey[z]

    rows_gvkey = ratio.loc[ratio['gvkey'] == b]                                                             ## Extract BS rows where 'stkcd' == b                                                                                           
    rows_date = rows_gvkey.loc[rows_gvkey['public_date'] < a].sort_values(by='public_date', ascending=True )  ## Extract subrows where annodt < a. Ascending sorting                                                                          

    if rows_date.size > 0:
        row_to_add = rows_date.tail(1)              ## Keep most recent entry, located at the end of the sorted list rows_annodt                                                                                                    
        dict_new_entries[z] = row_to_add.values[0]    ## Memorize this entry in dict                                                                                                                                                  

    else:
        # print('z:', z, ' has no entry: ')                                                                                                                                                                                           
        dict_new_entries[z] = empty_row               ## Empty row. Will generate NaNs in every columns                                                                                                                               

main = pd.DataFrame.from_dict(dict_new_entries, orient="index", columns=ratio.columns)   ## Convert dictionary into dataFrame                                                                                                            
print(main)
```

```python
#We concatenate main and yields together
complete=pd.concat([yields,main],axis=1)
complete = complete[pd.notnull(complete['public_date'])]  ## Get rid of the NaN rows
complete.to_pickle("./database.pkl")   ## We save the intermediary dataframe
```

```python
complete = pd.read_pickle("./database.pkl")
```

```python
SP500= pd.read_csv('^GSPC.csv',usecols=["Date", "Adj Close"])
SP500['sprtrn']= SP500['Adj Close'].pct_change()*100     ## Creates the column sprtrn with the monhtly SP500 return
SP500= SP500[np.isfinite(SP500['sprtrn'])]
SP500.Date=pd.to_datetime(SP500.Date)
complete = pd.merge(complete.assign(grouper=complete['datadate'].dt.to_period('M')),
               SP500.assign(grouper=SP500['Date'].dt.to_period('M')),
               how='left', on='grouper')
complete=complete.drop(columns=['grouper', 'Date','Adj Close'])
complete['win'] = np.where((complete.trt1m > complete.sprtrn) 
                     , 1, 0)
```

### Cleaning the dataset

```python
complete.isna().sum()
```

```python
## We cannot trade stocks with no monthly yields
clean_db = complete[np.isfinite(complete['trt1m'])]
## We can drop data that is not useful for us
clean_db=clean_db.drop(columns=['iid', 'tic','cusip','adate','qdate','public_date'])
```

```python
## We do not want to use items where more than 50% of the rows are missing. 
clean_db = clean_db.loc[:, clean_db.isnull().mean() < .5]
```

```python
## For the rest of the ratios, we will impute the missing values with the mean of the total column
imp=SimpleImputer(missing_values=np.nan, strategy="mean" )
#database1['ptb'] = imp.fit_transform(database1[['ptb']])
impute_columns=clean_db.columns[clean_db.isna().any()].tolist()
for i in impute_columns:
    clean_db[i] = imp.fit_transform(clean_db[[i]])
```

```python
## We can now visualize the cleaned DB:
clean_db.head()
```

### Feature Selection

```python
train = clean_db.drop(columns = ['win','conm','datadate','gvkey','trt1m'])
train_labels = clean_db['win']
fs = FeatureSelector(data = train, labels = train_labels)
```

```python
fs.identify_collinear(correlation_threshold=0.975)
fs.plot_collinear(plot_all=True)
```

```python
fs.identify_zero_importance(task = 'regression', eval_metric = 'auc', 
                            n_iterations = 10, early_stopping = True)
```

```python
fs.identify_low_importance(cumulative_importance = 0.99)
```

```python
all_to_remove = fs.check_removal()
print(all_to_remove)
clean_removed_db = clean_db.drop(columns = all_to_remove)
```

```python
database = clean_removed_db
rendement_mens=[]
rendement_mensSP=[]
for i in range(0,13):
    d = datetime.datetime.strptime("2015-12-31", "%Y-%m-%d")

    d2 = d - dateutil.relativedelta.relativedelta(months=i)
    d3 = d - dateutil.relativedelta.relativedelta(months=i+1)
    d4 = d3 - dateutil.relativedelta.relativedelta(years=7)    ##On impose un rolling window de 7 ans


    test_set = database[database["datadate"].isin(pd.date_range(d2, d2))]
    training_set=database[database["datadate"].isin(pd.date_range(d4, d3))]


    # Labels are the values we want to predict
    train_labels = np.array(training_set['win'])
    # Remove the labels from the features
    # axis 1 refers to the columns
    train_features= training_set.drop('win', axis = 1)
    train_features=train_features.drop('sprtrn',axis=1)
    train_features=train_features.drop('gvkey',axis=1)
    train_features=train_features.drop('datadate',axis=1)
    train_features=train_features.drop('conm',axis=1)
    train_features=train_features.drop('trt1m',axis=1)

    # Saving feature names for later use
    train_feature_list = list(train_features.columns)
    # Convert to numpy array
    train_features = np.array(train_features)


    test_labels = np.array(test_set['win'])
    # Remove the labels from the features
    # axis 1 refers to the columns
    test_features= test_set.drop('win', axis = 1)
    test_features=test_features.drop('datadate',axis=1)
    test_features=test_features.drop('sprtrn',axis=1)
    test_features=test_features.drop('trt1m',axis=1)
    test_features=test_features.drop('gvkey',axis=1)
    test_features=test_features.drop('conm',axis=1)

    # Convert to numpy array
    test_features = np.array(test_features)

    # Import the model we are using
    from sklearn.ensemble import RandomForestRegressor
    # Instantiate model with 1000 decision trees
    rf = RandomForestRegressor(n_estimators = 100, random_state = 42,n_jobs=20)
    # Train the model on training data
    rf.fit(train_features, train_labels);

    # Use the forest's predict method on the test data
    predictions = rf.predict(test_features)
    # Calculate the absolute errors
    predictions_actual = np.where((predictions > 0.50) 
                         , 1, 0)

    errors = abs(predictions_actual - test_labels)
    # Print out the mean absolute error (mae)
    print('Mean Absolute Error:', round(np.mean(errors),2))
    print('Sum of miss:',np.sum(errors))
    print('Fraction of miss:',np.sum(errors)/len(test_labels))


    test_return = np.array(test_set['trt1m'])
    Sp500_return=np.array(test_set['sprtrn'])
    porfolio=[]
    for i in range(0,len(predictions)):
        if predictions[i]>=0.75:
            porfolio.append(test_return[i])

    print("Le rendement du portfolio ce mois-ci est:",np.sum(porfolio)/len(porfolio))
    print("Le rendement du SP500 ce mois-ci est:",Sp500_return[0])
    
    rendement_mens.append(np.sum(porfolio)/len(porfolio))
    rendement_mensSP.append(Sp500_return[0])
```

```python
#print(np.sum(rendement_mens))
#print(np.sum(rendement_mensSP))

rendement_decim=np.divide(rendement_mens,100)
rendement_decimSP=np.divide(rendement_mensSP,100)
annualized_return=1
for i in range(0,len(rendement_decim)):
    annualized_return=annualized_return*(1+rendement_decim[i])
annualized_return=(annualized_return-1)*100
print("Rendement annuel composé du portefeuille:",np.round(annualized_return,2),"%")

annualized_returnSP=1
for i in range(0,len(rendement_decimSP)):
    annualized_returnSP=(annualized_returnSP*(1+rendement_decimSP[i]))
annualized_returnSP=(annualized_returnSP-1)*100
print("Rendement annuel composé du SP500",np.round(annualized_returnSP,2),"%")
```

```python
def rfstockpicker(database,prediction_date,rolling_window,nb_months,n_estimators,n_jobs,prediction_threshold):
    database = database
    rendement_mens=[]
    rendement_mensSP=[]
    for i in range(0,nb_months):
        d = datetime.datetime.strptime(prediction_date, "%Y-%m-%d")

        d2 = d - dateutil.relativedelta.relativedelta(months=i)
        d3 = d - dateutil.relativedelta.relativedelta(months=i+1)
        d4 = d3 - dateutil.relativedelta.relativedelta(years=rolling_window)    ##On impose un rolling window de 7 ans

        test_set = database[database["datadate"].isin(pd.date_range(d2, d2))]
        training_set=database[database["datadate"].isin(pd.date_range(d4, d3))]


        # Labels are the values we want to predict
        train_labels = np.array(training_set['win'])
        # Remove the labels from the features
        # axis 1 refers to the columns
        train_features= training_set.drop('win', axis = 1)
        train_features=train_features.drop('sprtrn',axis=1)
        train_features=train_features.drop('gvkey',axis=1)
        train_features=train_features.drop('datadate',axis=1)
        train_features=train_features.drop('conm',axis=1)
        train_features=train_features.drop('trt1m',axis=1)

        # Saving feature names for later use
        train_feature_list = list(train_features.columns)
        # Convert to numpy array
        train_features = np.array(train_features)
        test_labels = np.array(test_set['win'])
        # Remove the labels from the features
        # axis 1 refers to the columns
        test_features= test_set.drop('win', axis = 1)
        test_features=test_features.drop('datadate',axis=1)
        test_features=test_features.drop('sprtrn',axis=1)
        test_features=test_features.drop('trt1m',axis=1)
        test_features=test_features.drop('gvkey',axis=1)
        test_features=test_features.drop('conm',axis=1)

        # Convert to numpy array
        test_features = np.array(test_features)

        # Import the model we are using
        from sklearn.ensemble import RandomForestRegressor
        # Instantiate model with 1000 decision trees
        rf = RandomForestRegressor(n_estimators = n_estimators, random_state = 42,n_jobs=n_jobs)
        # Train the model on training data
        rf.fit(train_features, train_labels);

        # Use the forest's predict method on the test data
        predictions = rf.predict(test_features)


        test_return = np.array(test_set['trt1m'])
        comp_name=np.array(test_set['conm'])
        Sp500_return=np.array(test_set['sprtrn'])
        porfolio=[]
        compname_mens=[]
        for i in range(0,len(predictions)):
            if predictions[i]>=prediction_threshold:
                porfolio.append(test_return[i])
                compname_mens.append(comp_name[i])
                
        print("Le rendement du portfolio ce mois-ci est:",np.sum(porfolio)/len(porfolio))
        print("Le rendement du SP500 ce mois-ci est:",Sp500_return[0])
        print(compname_mens)

        rendement_mens.append(np.sum(porfolio)/len(porfolio))
        rendement_mensSP.append(Sp500_return[0])
        
    return rendement_mens,rendement_mensSP,
        
```
